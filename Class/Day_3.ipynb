### MSE, RMSE 계산 실습

```python
import tensorflow as tf
import numpy as np

x = [2, 4, 6, 8]
y = [81, 93, 91, 97]

a = 3 # 기울기
b = 76 # y절편

prediction = [(a * i) + b for i in x]

print(f"실제 값(y): {y}")
print(f"예측 값(p): {prediction}")
print("==================")

loss = [(prediction[i] - y[i]) for i in range(len(x))]
mse = np.mean([l**2 for l in loss])
rmse = np.sqrt(mse)

print(f"오차 값: {loss}")
print(f"MSE 값: {mse}")
print(f"RMSE 값: {rmse}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform

# 한글 폰트
font_manager.fontManager.addfont('/usr/share/fonts/truetype/nanum/NanumGothic.ttf')
plt.rcParams['font.family'] = 'NanumGothic'
# 마이너스 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
def f(x):
    return x**2
def df(x):
    return 2*x

x = 3.0
alpha = 0.9
num_iterations = 50

x_history = [x]
f_history = [f(x)]
print(f"초기값: x = {x:.4f}, f(x) = {f(x):.4f}")

for i in range(num_iterations):
    gradient = df(x)
    x -= alpha * gradient
    x_history.append(x)
    f_history.append(f(x))

    print(f"iteration {i+1}: x = {x:.4f}, gradient = {gradient:.4f}, f(x) = {f(x):.4f}")

x_vals = np.linspace(-3, 3, 100)
y_vals = f(x_vals)
plt.figure(figsize=(7, 10))
plt.plot(x_vals, y_vals, label="f(x) = x^2")
plt.plot(x_history, f_history, '.-', label="경사하강 이동경로")

plt.title("경사하강법으로 x^2 최소화하는 과정")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = [[2, 81], [4, 93], [6, 91], [8, 97]]

x = [i[0] for i in data]
y = [i[1] for i in data]

plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.show()

x_data = np.array(x)
y_data = np.array(y)

a = 0
b = 0

lr = 0.03 # learning rate에 따라 정확한 계산식임에도 값이 제대로 구해지지 않을 수 있음. 가장 효과적인 lr은 0.03
epochs = 2001

for i in range(epochs):
    y_pred = a * x_data + b
    error = y_data - y_pred
    a_diff = 2 / len(x_data) * sum((a * x_data + b - y_data) * x_data)
    b_diff = 2 / len(x_data) * sum(a * x_data + b - y_data)

    a -= lr * a_diff
    b -= lr * b_diff

    if i % 100 == 0:
        print(f"epoch : {i}, 기울기 : {a}, 절편 : {b}")

y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```
