### input() 함수
input 함수 괄호 안에 텍스트를 넣어 입력받을 수 있다.

```python
a = input("값을 입력하세요: ")
print(a)
```

```python
name = input("이름")
age = int(input("나이"))
print("이름:", name, "나이:", age, sep=" ")
```

```python
num1, num2 = map(int, input("값을 입력하세요:").split())
print(num1 + num2)
```

```python
num1, num2, num3 = map(int, input("값을 입력하세요:").split())
print((num1 + num2 + num3) / 3)
```

```python
score = int(input("점수를 입력하세요 : "))
if 100 >= score >= 90:
    print("A")
elif score < 90 and score >= 80:
    print("B")
elif score < 80 and score >= 70:
    print("C")
elif score < 70 and score >= 60:
    print("D")
elif score < 60 and score >= 0:
    print("F")
else:
    print("다시 입력하세요.")
```

```python
def multi(num1, num2, num3):
    return num1 * num2 * num3

num1, num2, num3 = map(int, input('숫자 3개를 입력하세요(예: 1,2,3): ').split(','))
print("숫자 3개의 곱:", multi(num1, num2, num3))
```

```python
def operator(num1, num2):
    return num1+num2, num1-num2, num1*num2

num1, num2 = map(int, input('숫자 2개를 입력하세요(예: 1,2): ').split(','))
result1, result2, result3 = operator(num1, num2)
print("숫자 2개의 합:", result1, "숫자 2개의 차:", result2, "숫자 2개의 곱:", result3)
```
